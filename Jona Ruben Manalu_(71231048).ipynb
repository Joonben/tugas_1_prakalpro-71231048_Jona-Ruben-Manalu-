```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,'b',x,z,'r')
plt.xlabel('Radians');
plt.ylabel('velue');
plt.title('Plotting Demonstration')
plt.legend(['Sin','Cos'])
plt.grid()
```

```python
#harga emas saat ini
harga_emas_sekarang = 685000

#harga emas awal
harga_emas_awal = 650000

#jumlah gram yang dibeli gerard
jumlah_gram_emas_awal = 25

#jumlah gram yang dibeli lagi oleh gerard
jumlah_gram_emas_lagi = 15

#menghitung keuntungan dari pembelian emas awal
keuntungan_25_gram = (harga_emas_sekarang - harga_emas_awal) * jumlah_gram_emas_awal

#menghitung keuntungan dari pemebelian emas sekarang
keuntungan_15_gram = (harga_emas_sekarang - harga_emas_awal) * jumlah_gram_emas_lagi

#menghitung total keuntungan yang di dapat oleh gerard
total_keuntungan = keuntungan_25_gram + keuntungan_15_gram

#menghitung peresentase keuntungan yang di dapatkan oleh gerard
persentase_keuntungan = (total_keuntungan / (harga_emas_awal * jumlah_gram_emas_awal + harga_emas_sekarang * jumlah_gram_emas_lagi ) ) * 100

#menampilkan hasil akhir atau keuntungan dan persentasnya
print("Keuntungan dari pembelian emas 25 gram : Rp.", keuntungan_25_gram)
print("Keuntungan dari pembelian emas lagi 15 gram : Rp.",keuntungan_15_gram)
print("Total keuntungan yang didapat oleh Gerard : Rp.", total_keuntungan)
print("Persentase keuntungan yang didapat oleh Greard : ", persentase_keuntungan, "%")
```

```python
Awal = 200000000
Akhir = 400000000
Bunga = 0.1
n = 1
Tahun = 0

while Awal <= Akhir:
    Awal= Awal * (1+Bunga/n)
    Tahun += 1
    print("Uang Erika sesudah", Tahun, "Tahun : Rp.", round(Awal))
    
print("Untuk mencapai target Rp.400jt, maka di butuhkan ", Tahun, "Tahun")
```
