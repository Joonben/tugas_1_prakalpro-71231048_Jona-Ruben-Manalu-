```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,'b',x,z,'r')
plt.xlabel('Radians');
plt.ylabel('velue');
plt.title('Plotting Demonstration')
plt.legend(['Sin','Cos'])
plt.grid()
```

```python
#harga emas saat ini
harga_emas_saat_ini = 685000

# harga emas sebelumnya
harga_emas_sebelumnya = 650000

#jumlah gram emas yang dibeli gerard
jumlah_gram_emas = 25

#jumlah gram emas yang dibeli lagi gerard
jumlah_gram_emas_lagi = 15

# Menghitung keuntungan dari pembelian 25 gram emas
keuntungan_25_gram = (harga_emas_saat_ini - harga_emas_sebelumnya) * jumlah_gram_emas

# Menghitung keuntungan dari pembelian 15 gram emas
keuntungan_15_gram = (harga_emas_saat_ini - harga_emas_sebelumnya) * jumlah_gram_emas_lagi

# Menghitung total keuntungan yang didapat oleh Gerard
total_keuntungan = keuntungan_25_gram + keuntungan_15_gram

# Menghitung persentase keuntungan yang didapat oleh Gerard
persentase_keuntungan = (total_keuntungan / (harga_emas_sebelumnya * jumlah_gram_emas + harga_emas_saat_ini * jumlah_gram_emas_lagi)) * 100

# Menampilkan hasil keuntungan dan persentase keuntungan
print("Keuntungan dari pembelian 25 gram emas: Rp. ", keuntungan_25_gram)
print("Keuntungan dari pembelian 15 gram emas: Rp. ", keuntungan_15_gram)
print("Total keuntungan yang didapat oleh Gerard: Rp. ", total_keuntungan)
print("Persentase keuntungan yang didapat oleh Gerard: ", persentase_keuntungan, "%")
```
